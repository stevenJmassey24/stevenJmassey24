```python
#hypothesis function for the single variable LR
def hypothesis(theta_array,x):
  return theta_array[0]+theta_array[1]*x
```

```python
def costfunction(theta_array,x_value,y_value,m):
  total_error=0
  for i in range(m):
    total_error+=(theta_array[0]+theta_array[1]*x_value[i] - y_value[i])**2
  return total_error/(2*m)
```

```python
def gradient_Descent(theta_array,X,Y,alpha,m):
  summation_1=0
  summation_0=0

  for i in range(m):
    #theta_0
    summation_0+=(theta_array[0]+theta_array[1]*X[i])-Y[i]
    #theta_1
    summation_1+=X[i]*((theta_array[0]+theta_array[1]*X[i])-Y[i])
  new_theta0=theta_array[0]-alpha*(summation_0)/m
  new_theta1=theta_array[1]-alpha*(summation_1)/m

  improved_theta=[new_theta0,new_theta1]
  print(improved_theta)
  return improved_theta
```

```python
def training(x_train,y_train,alpha,iters):
  m=x_train.size
  theta_0=0
  theta_1=0
  theta_array=[theta_0,theta_1]
  cost_values=[]
  #goal is to find theta_array:gradientDescent
  for i in range(iters):
    theta_array=gradient_Descent(theta_array,x_train,y_train,alpha,m)
  #to minimize the cost values: costfunction
    loss=costfunction(theta_array,x_train,y_train,m)
    cost_values.append(loss)

    ynew=theta_array[0]+theta_array[1]*x_train
    plt.plot(x_train,ynew,'r')
    plt.scatter(x_train,y_train)
    plt.show()
  x=np.arange(0,iters,step=1)
  plt.plot(x,cost_values,"-b")
  plt.xlabel("Number of iterations")
  plt.ylabel("cost values")
  plt.title("Cost function curve")
  plt.show()
  return theta_array
```
