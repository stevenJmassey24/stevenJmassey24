```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel('kmean.xlsx', sheet_name='ex7data2_X', header=None)
df1.head()
```

```python
import random #takes random constant values
indexes=random.sample(range(0,len(df1)),3)
```

```python
indexes
```

```python
[]
```

```python
centroids = []
for i in indexes:
    centroids.append(df1.loc[i])
centroids
```

```python
centroids = np.array(centroids)
```

```python
centroids
```

```python
def distance_calc(X1,X2):
 return(sum((X1-X2)**2))**0.5
```

```python
def findClosestCentroids(c, X):
    assigned_centroid = []
    for i in X:
        distance=[]
        for j in c:
            distance.append(distance_calc(i, j)) #centroids
        assigned_centroid.append(np.argmin(distance))
    return assigned_centroid
```

```python
get_centroids = findClosestCentroids(centroids, x)
get_centroids
```
